```python
x =df.drop("Outcome",axis=1)
y =df["Outcome"]
```

```python





x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# numpy array to pytorch tensors
x_train_tensor= torch.from_numpy(x_train)
x_test_tensor= torch.from_numpy(x_test)
x_train_tensor= torch.tensor(x_train,dtype=torch.float32)
x_test_tensor= torch.tensor(x_test,dtype=torch.float32)

y_train_tensor= torch.tensor(y_train.values,dtype=torch.float32)
y_test_tensor= torch.tensor(y_test.values,dtype=torch.float32)
```

```python
#Define  the  nn model
class MyNN(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.model = nn.Sequential(
        nn.Linear(num_features,8),
        nn.ReLU(),
        nn.Linear(8,4),
        nn.ReLU(),
        nn.Linear(4,1),
        nn.Sigmoid()

  )

  def forward(self,features):
    return self.model(features)

model=MyNN(x_train_tensor.shape[1])
```

```python
#define the loop

for epoch in range(epochs):
  #forward pass
  y_pred = model(x_train_tensor)
  #Loss calculate
  loss = loss_function(y_pred,y_train_tensor.reshape(-1,1))

  #zero graadients

  optimizer.zero_grad()

  # backward pass

  loss.backward()

  # parameters upgrade

  optimizer.step()


  #print loss in each epoch
  if (epoch+1)%10==0:
     print(f"Epoch :{epoch},Loss :{loss.item()}")
```

#Answer to the qus no 1:

WE know that in pytorch gradients are accumulated by default ,when we call loss.backward() the computed gradients are added to the existing gradients store in the parameters  

If we dont reset the gradients before the next epoch or iteration the new gradients will be added to the previous ones. This causes incorrect weight updates and can lead to unstable trainig

There fore gradients must be reset
